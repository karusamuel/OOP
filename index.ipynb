
# Programming Paradigms

Programming paradigms are different styles of structuring code.

### 1. Procedural Programming
- Code is written as a sequence of steps
- Uses functions and control flow (if statements, loops)
- Common in simple scripts and data science notebooks

### 2. Functional Programming
- Focuses on functions and function composition
- Avoids changing state and global variables
- Used in some patterns within libraries such as Matplotlib

### 3. Object-Oriented Programming (OOP)
- Organizes code into objects with data and behavior
- Helps structure large and complex systems
- Improves code organization and scalability

## Introduction to Object-Oriented Programming (OOP)

Object-Oriented Programming (OOP) is a programming paradigm that structures code around objects, which are instances of classes. Each object contains:

- Attributes → data or properties (e.g., name, age)
- Methods → functions or behaviors (e.g., run(), calculate())

OOP is widely used in languages such as Python, Java, C++, and C#, and is designed to improve modularity, reusability, and maintainability of code.

# Types of Attributes in Python (OOP)

In Object-Oriented Programming (OOP), attributes are variables that belong to a class or an object. They define the state or properties of objects.

Python has three main types of attributes:


## 1. Instance Attributes

Instance attributes are attributes that belong to a specific object created from a class. Each object has its own independent copy of these attributes, meaning changes to one object do not affect another. They are typically used to store data that is unique to each instance.

```python
# example with instance attribute 
class Student:
    def __init__(self,name,phone):
        self.name = name
        self.phone = phone
        self.cohort = "DSF-FT-16"
```

```python
sam = Student("sam karu","0712 xxx xxx")
```

```python
sam.cohort = "DSF-PT-16"
```

```python
sam.cohort
```


## 2. Class Attributes

Class attributes are attributes that belong to the class itself rather than individual objects. They are shared across all instances of the class. If the value of a class attribute is changed, it affects all objects that access it unless it is overridden at the instance level.


```python
# example
class Student:
    # class attribute
    cohort = "DSF-FT-16"
    # constructer / instance method
    def __init__(self,name,phone):
        #  instance attributes 
        self.name = name
        self.phone = phone
```

```python
# example usage 
sam = Student("samuel karu","0714 xxx xxx")
```

```python
# print Class attr
sam.cohort
```

## 3. Private Attributes

Private attributes are attributes intended to be hidden from direct access outside the class. They are used to protect internal data and enforce encapsulation. While they can still be accessed indirectly in Python, they are designed to discourage external modification and promote controlled access through class methods.


```python
# example
class Student:
    # class attribute
    cohort = "DSF-PT-15"
    # constructer / instance method
    def __init__(self,name,phone,password):
        #  intance attributes 
        self.name = name
        self.phone = phone
        self.__password = password
```

```python
# usage 

sam = Student("sam karu","0714 xxx xxx","qwerty")
```

```python
sam.name
```

```python
sam.phone
```

```python
# access private attributes 
# sam.__password
```

## Summary 

| Type               | Defined Where       | Scope        | Shared? | Access Level |
| ------------------ | ------------------- | ------------ | ------- | ------------ |
| Instance attribute | Inside `__init__`   | Object level | No      | Public       |
| Class attribute    | Inside class body   | Class level  | Yes     | Public       |
| Private attribute  | Inside class (`__`) | Object level | No      | Restricted   |


## Instance Methods

Instance methods are functions defined inside a class and are bound to objects created from that class.

## Methods vs Attributes

An instance object can have both attributes and methods:

### Attributes
- Store data or properties
- Example: name, age, color
- Contain information about the object

### Methods
- Are functions defined inside a class
- Perform actions or behavior
- Are callable (they run code when called)


```python
# example
class Student:
    # class attribute
    cohort = "DSF-PT-15"
    # constructer / instance method
    def __init__(self,name,phone):
        #  intance attributes 
        self.name = name
        self.phone = phone
        
    def send_message(self,message):
        print("hello",self.name)
        print(message)
```

```python
# innitialize
sam = Student(name="Samuel Karu",phone="0714 xxx xxx")
```

```python
# call methods
sam.send_message("class has been moved to 2:00 pm")
```

```python
brian = Student(name="Brian kipkemoi",phone="0714 ccc ccc")
```

```python
brian.send_message("class has been moved to 2:00 pm")
```


## What is Inheritance?


Inheritance allows you to create a **new class based on an existing class**.

### Key idea:
- The existing class is called the **parent (or base) class**
- The new class is called the **child (or derived) class**
- The child class automatically gets the attributes and methods of the parent class


## Why Inheritance is Useful

Inheritance helps to:

- Reuse existing code
- Avoid repetition
- Extend functionality without rewriting everything
- Organize code in a clear hierarchy

## The DRY Principle

DRY stands for:

> Don't Repeat Yourself

You should avoid duplicating code unnecessarily.

Without inheritance:
- You may rewrite the same attributes and methods in multiple classes
- This leads to repeated and harder-to-maintain code

With inheritance:
- Shared logic is written once in the parent class
- Child classes reuse it automatically
- Only differences are added or overridden

```python
# Base User Class 
class User:
    
    def __init__(self,name,phone,email):
        self.name = name
        self.phone = phone
        self.email = email
        
    def send_message(self,message):
        print("hello",self.name)
        print(message)
```

```python
# class student

class Student(User):
    
    def __init__(self, name, phone, email,student_email):
        super().__init__(name, phone, email)
        self.student_email = student_email
        
    def mark_attedance(self,present=True):
        if present:
            print(self.name,"was marked as present")
        else:
            print(self.name,"was marked as absent")
```

```python
# class Tm 
class TM(User):
    
    def __init__(self, name, phone, email,work_email):
        super().__init__(name, phone, email)
        self.work_email = work_email
        
    def assign_student(self,student:Student):
        print(student.name,"was assigned to",self.name)
```

## Example Usage

```python
# example usage with student 

sam = Student(name="samuel karu",phone="0714 xxx xxx",email="cccc@gmail.com",student_email="cccc@student.com")

sam.send_message("you were assigned a class")
```
